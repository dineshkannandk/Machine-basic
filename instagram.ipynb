Highest number of instagram influencers

```python
f=pd.read_csv('influencer.csv')
f.dropna(inplace=True)
xi=f['Followers'].str.replace('m','').astype(float).value_counts()
yi=f['Country Or Region'].value_counts()
yi
sns.barplot(x=xi.index,y=xi.values,data=f)
sns.barplot(x=yi.index,y=yi.values,data=f)
```

```python
#Followers and likes count
f=pd.read_csv('influencer.csv')
x=f['Followers'].str.replace('m','')
y=f['Avg. Likes'].str.replace('m','')
sns.lineplot(x=x,y=y,data=f)
plt.xlim(1, 10)  
plt.ylim(1, 10)
```
